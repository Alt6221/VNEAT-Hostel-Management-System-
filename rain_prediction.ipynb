```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Wind.csv')
df
```

```python
from sklearn import preprocessing
sti = preprocessing.LabelEncoder()
df = df.apply(sti.fit_transform)
df
```

```python
featured_cols = ['Rainfall', 'Evaporation', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RainToday', 'RISK_MM']
x = df[featured_cols]
y = df.RainTomorrow
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor
rain = DecisionTreeRegressor(random_state=100)
rain.fit(x_train, y_train)
```

```python
y_pred = rain.predict(x_test)
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
result
```

```python
from sklearn.metrics import accuracy_score
print("Acurracy: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
